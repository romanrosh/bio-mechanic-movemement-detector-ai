```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Most Squats in 5-Minutes.csv')
```

```python
df.columns
```

```python
head_x = (df.loc[:,['0-XNose','15-XREye','16-XLEye','17-XREar','18-XLEar']])
```

```python
head_y = (df.loc[:,['0-YNose','15-YREye','16-YLEye','17-YREar','18-YLEar']])
head_y
```

```python
df['XHead']=head_x.apply(np.mean, axis=1)
```

```python
df['YHead']=head_y.apply(np.mean, axis=1)
```

```python
# df.loc[:,['XHead', 'YHead']]
display(df.YHead)
```

```python
plt.plot(df['XHead'])
```

```python
plt.plot(df['YHead'])
```

```python
plt.plot(df['YHead'][:400])
```

```python
plt.plot(df['YHead'][:350])
```

```python
clean_df = df.iloc[:350, :]
```

```python
labels = []
df['label'] = np.zeros(df.shape[0])
length = df.shape[0]
for index in np.arange(1, length-1):
    if df['YHead'][index]>df['YHead'][index-1] and df['YHead'][index]>df['YHead'][index+1]:
        df.label[index] = 'bottom'
    elif df['YHead'][index]<df['YHead'][index-1] and df['YHead'][index]<df['YHead'][index+1]:
        df.label[index] = 'top'
    else:
        df.label[index] ='midway'
```

```python
df = df.iloc[:-1]
```

```python
print(df.label.tail(15))
plt.plot(df.YHead.tail(100))
```

```python
for i in np.arange(200, df.shape[0]):
    print(f"height: {df.YHead[i]}, label: {df.label[i]}")
```
